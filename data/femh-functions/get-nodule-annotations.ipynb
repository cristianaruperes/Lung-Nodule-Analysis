```python
import pandas as pd
import shutil
import os

# Load the CSV file into a DataFrame
csv_file = 'dataset/fix-data_CSV.csv'  # Replace with the path to your CSV file
df = pd.read_csv(csv_file)

prefix = 'dataset/18aug-fix-dataset/'  # Replace with the desired prefix

# Define the destination directory where you want to move the matching files
destination_directory = 'dataset/filtered_files_with_nodule'  # Replace with your desired directory path

# Iterate through the DataFrame and filter files within the directories
for index, row in df.iterrows():
    directory_path = prefix + str(row['Path'])  # Replace 'Path' with the actual column name
    unique_id = row['Index']  # Replace 'Unique_id' with the actual column name

    # Check if the directory exists
    if os.path.exists(directory_path):
        # Iterate through the files in the directory
        for filename in os.listdir(directory_path):
            # Check if the filename contains "Nodule," "nodules," or "nodule" (case-insensitive)
            if any(keyword.lower() in filename.lower() for keyword in ["Nodule", "nodules", "nodule"]):
                # Construct the full source path to the file
                source_path = os.path.join(directory_path, filename)
                print(f'Moving file: {source_path}')

                # Construct the new filename using the unique ID
                new_filename = f'{unique_id}_{filename}'  # You can customize the naming pattern as needed

                # Construct the full destination path for the file
                destination_path = os.path.join(destination_directory, new_filename)

                # Create any missing directories in the destination path
                os.makedirs(os.path.dirname(destination_path), exist_ok=True)

                # Move the file to the destination directory with the new filename
                shutil.move(source_path, destination_path)

print(f'Files containing "Nodule," "nodules," or "nodule" moved to: {destination_directory}')
```
