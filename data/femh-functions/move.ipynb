```python
import pandas as pd
import shutil
import os

# Load the CSV file into a DataFrame
csv_file = 'dataset/fix-data_CSV.csv'  # Replace with the path to your CSV file
df = pd.read_csv(csv_file)

prefix = 'dataset/18aug-fix-dataset/'   # Replace with the desired prefix

# Define the destination directory where you want to save the images
destination_directory = 'dataset/fix_data_images'  # Replace with your desired directory path

# Iterate through the DataFrame and copy/move images to the destination directory
for index, row in df.iterrows():
    # directory_path = prefix + row['Path']  # Replace 'Directory' with the actual column name
    directory_path = prefix + row['Path']  # Replace 'Directory' with the actual column name
    # print(directory_path)
    filename = str(row['Filename'])  # Replace 'Filename' with the actual column name
    unique_id = row['Index']  # Replace 'ID' with the actual column name containing unique IDs

    # Construct the full source path to the image
    source_path = directory_path + '/' + filename
    print(source_path)
    
    # Get the file extension from the original filename
    file_extension = os.path.splitext(filename)[1]
    
    # Construct the new filename using the unique ID
    new_filename = f'{unique_id}{file_extension}'
    
    # Construct the full destination path for the image
    destination_path = os.path.join(destination_directory, new_filename)
    
    # Copy or move the image to the destination directory (use shutil.copy or shutil.move)
    shutil.copy(source_path, destination_path)  # Use shutil.move if you want to move instead of copy

print(f'Images copied/moved to: {destination_directory}')
```
